```python
import math
prm_nmbr=[]
for nmbr in range(10,101):
    for i in range(2,int(math.sqrt(nmbr))):
        if nmbr%i==0:
            break
    else:
        prm_nmbr.append(nmbr)
print(prm_nmbr)
for x in prm_nmbr:
    for y in prm_nmbr:
        for z in prm_nmbr:
            if x!=y!=z and (x+y+z)/3==y:
                print(x,y,z)
```

```python
def correct_stream(user, correct):
	liste = []
	for i,j in zip(user,correct):
		if i==j:
			liste.append(1)
		else:
			liste.append(-1)
	return liste
correct_stream(
  ["it", "is", "find"],
  ["it", "is", "fine"]
)
```

```python
def correct_stream(user, correct):
    return [1 if i==j else -1 for i,j in zip(user,correct)]
correct_stream(
  ["it", "is", "find"],
  ["it", "is", "fine"]
)
```
